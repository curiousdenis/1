```python
import pandas as pd
import matplotlib.pyplot as plt
import os 
```

```python
pwd = os.getcwd()#cwd - current working directory --> just to make sure that everything has come from current directory
```

```python
filename = "\\Data - Survey Monkey Output.xlsx"
dataset = pd.read_excel(pwd + filename, sheet_name = "second_step" )
```

```python
# let's make a copy of dataset, in case we screw it and you would not import again 
dataset_copy = dataset.copy()
```

```python
#dataset_copy.columns --> check all the columns 
columns_to_drop = ['Start Date', 'End Date', 'Email Address',
       'First Name', 'Last Name', 'Custom Data 1']
dataset_copy = dataset_copy.drop(columns = columns_to_drop)
dataset_copy
```

```python
# create argumanets for melt function()
id_vars = list(dataset_copy.columns[:8])
value_vars = list(dataset_copy.columns[8:])
```

```python
# melt - meaning make a horisontal column vertical 
dataset_copy_melted = dataset_copy.melt(id_vars = id_vars , value_vars = value_vars, 
                                        var_name = "Question + Subquestion" , value_name = "Answer")
dataset_copy_melted
```

```python
# make a new sheet of Question so I can add to dataset_copy_melted
questions = pd.read_excel(pwd + filename, sheet_name = "first_step")
questions_copy = questions.copy()
questions_copy
```

```python
# make a new sheet of Question so I can add to dataset_copy_melted
questions_to_drop = ["Main_header_raw" , "Subquestions_raw" , "Subquestions" , 
                     "Main_header_Subquestions_ValueOnly"]
question_copy = questions_copy.drop(columns = questions_to_drop , inplace = True)
```

```python
questions_copy
```

```python
# merge a new column which is Questions
dataset_merged = pd.merge(left = dataset_copy_melted  , right = questions_copy
                          , how = "left" , left_on = "Question + Subquestion" , right_on = "Question + Subquestion")
print(f"{len(dataset_merged)} - rows in dataset_merged")
print(f"{len(dataset_copy_melted)} - rows in dataset_copy_melted")
new_columns_name = ['Respondent ID', 'Identify which division you work in. - Response',
       'Identify which division you work in. - Other (please specify)',
       'Which of the following best describes your position level? - Response',
       'Which generation are you apart of? - Response',
       'Please select the gender in which you identify. - Response',
       'Which duration range best aligns with your tenure at your company? - Response',
       'Which of the following best describes your employment type? - Response','Questions',
       'Question + Subquestion', 'Answer']
dataset_merged = dataset_merged.reindex(columns = new_columns_name)
```

```python
# create a new dataframe where there is no NA answers --> so you can count QN + Unique ID
respondents = dataset_merged[dataset_merged["Answer"].notna()]
```

```python
# add new column, which is Total Respondents --> the number of people who gave answer to the Q1 - 
#basically skip all people people who gave no answer and count only those who did answer {QN + Unique ID}
# Q1 5379192392  - no answer (skip)  ,Q1 2658722536  - Answer 4 .....
respondents = respondents.groupby("Questions")["Respondent ID"].nunique().reset_index()
respondents.rename(columns = {"Respondent ID" : "Total answer"} , inplace = True)
```

```python
dataset_merged_two = pd.merge(left = dataset_merged  , right = respondents
                          , how = "left" , left_on = "Questions" , right_on = "Questions")
```

```python
dataset_merged_two
```

```python
# now I wanna find how many people give the same answer per question
# Q1 - Answer 4 
same_answer = dataset_merged_two.groupby(["Question + Subquestion" , "Answer"])["Respondent ID"].count().reset_index()
same_answer.rename(columns = {"Respondent ID" : "Same Answer"} , inplace = True)
```

```python
#merge dataset_merge_two + same_answer 
dataset_merged_three = pd.merge(left = dataset_merged_two  , right = same_answer
                          , how = "left" , left_on = ["Question + Subquestion" , "Answer"]
                               , right_on = ["Question + Subquestion" , "Answer"])
dataset_merged_three["Same Answer"].fillna(0, inplace = True)
output = dataset_merged_three.copy()
```

```python
output.to_excel(pwd + "\\Final_output.xlsx" , index = False)
```
